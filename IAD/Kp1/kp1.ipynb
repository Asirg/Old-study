 <h1 align="center" style="color:#4ca8ed">1. Бінаризація даних.</h1>

```python
# Бінаризація або розбиття на інтервали за допомогою лямбда виразу
def BinarLam(data, arg, func, name = ""):
    name = name if (name != "") else 'BinarLam ' + arg
    data[name] = data[arg].apply(func)
# Бінаризація або розбиття на інтервали за допомогою массиву зі значеннями границь интервалів
def Binar(data, arg, inter, name = ""):
    name = name if (name != '') else 'Binar ' + arg
    data[name] = data[arg]
    for i in range(len(data[name])):
        r_bord = True
        for j in range(len(inter)):
            if(inter[j] > data.loc[i,name]):
                data.loc[i,name] = j
                r_bord = False
                break
        if(r_bord):
            data.loc[i,name] = len(inter)
```

```python
BinarLam(data,'Age',lambda x: 0 if x > 30 else 1)
BinarLam(data,'Age',lambda x: 0 if x > 30 else (1 if x > 20 else 2),"BinarLam 2")
Binar(data,'Fare',[10,40,70])
data
```

 <h1 align="center" style="color:#4ca8ed">2. Кодування текстових даних</h1>

```python
# Заміна унікальних значеннь, на числа натурального ряду
def Coding(data,arg,name=""):
    name = name if (name != '') else 'Coding ' + arg
    uniq = []
    for i in data[arg]:
        if (not i in uniq):
            uniq.append(i)
    data[name] = [uniq.index(i) for i in data[arg]]
    print({x:y for x,y in zip(uniq,range(len(uniq)))})
```

```python
Coding(data,'Embarked')
data
```

 <h1 align="center" style="color:#4ca8ed">3. Заповнення пропущенних значень</h1>

```python
# Заповнення пропусків деяким значенням, в залежності від вибору
def FillGapsNum(data, arg, tp, ind = 0, name = "",val = 0):
    name = name if (name != '') else 'FillGapsNum '+ str(tp) + ' ' + arg
    if (tp == 0):
        val = 0
    elif (tp == 1):
        val = data[arg].mean()
    elif (tp == 2):
        val = data[arg].median()
    elif (tp == 3):
        val = data[arg].mode()[0]
    data[name] = [i if(i == i) else val for i in data[arg]]
    if (ind == True and not 'Indicator ' + arg in data.columns):
        data['Indicator ' + arg] = [0 if(i == i) else 1 for i in data[arg]]
# Заповнення пропусків останнім значенням, або ближнім
def FillGapsLast(data, arg, ind = 0, name = ""):
    name = name if (name != '') else 'FillGapsLast ' + arg
    data[name] = data[arg]
    if (not data.loc[0,name] == data.loc[0,name]):
        for i in data[name]:
            if (i == i):
                data.loc[0,name] = i
                break
    data[name] = [data[name][i] if(data[name][i] == data[name][i]) else data[name][i-1] 
                  for i in range(len(data[name]))]
    if (ind == True and not 'Indicator ' + arg in data.columns):
        data['Indicator ' + arg] = [0 if(i == i) else 1 for i in data[arg]]
```

```python
FillGapsNum(data,'Age',0,1,'0')
FillGapsNum(data,'Age',1,1,'mean')
FillGapsNum(data,'Age',2,1,'median')
FillGapsNum(data,'Age',3,1,'mode')
FillGapsNum(data,'Age',4,1,'100',100)

FillGapsLast(data,'Age',0)
data.head(7)
```

 <h1 align="center" style="color:#4ca8ed">4. Процедури нормалізації даних</h1>

```python
def DnormalMM(data,arg,name=''):
    name = name if (name != '') else 'DnormalMM ' + arg
    xmin = data[arg].min()
    xmax = data[arg].max()
    data[name] = [(x-xmin)/(xmax-xmin) for x in data[arg]]

def DnormalMM2(data,arg,name=''):
    name = name if (name != '') else 'DnormalMM2 ' + arg
    xmin = data[arg].min()
    xmax = data[arg].max()
    data[name] = [2*(x-xmin)/(xmax-xmin)-1 for x in data[arg]]

def DnormalZ(data,arg,name=''):
    name = name if (name != '') else 'DnormalMMZ ' + arg
    std = data[arg].std()
    mean = data[arg].mean()
    data[name] = [(x-mean)/std for x in data[arg]]
```

```python
data.loc[0,'Fare'] = data['Fare'].min()
data.loc[1,'Fare'] = data['Fare'].max()
DnormalMM(data,'Fare')
DnormalMM2(data,'Fare')
DnormalZ(data,'Fare')
data
```
