<h1 style="color:#4ca8ed">2. Реалізувати алгоритм знаходження порогового значення за допомогою ROC-аналізу.</h1>
<h1 style="color:#4ca8ed">3. Вивести  таблицю значень чутливості та специфічності для кожного можливого порового значення.</h1>

<h1 style="color:#4ca8ed">5. Вивести графік ROC-кривої та розрахувати прощу під нею.</h1>

```python
def ROC(data, method):
    data =  data.sort_index().sort_values(0, kind='mergesort')
    data['поріг'] = [ 0 for i in range(len(data))]
    data['se'] = [ 0 for i in range(len(data))]
    data['sp'] = [ 0 for i in range(len(data))]
    
    p = sum(data['табл значень'])
    n = sum([1 if(i == 0) else 0 for i in data['табл значень']])
    Tp = 0; Tn = n
    print('Одиниці = ' + str(p))
    print('Нулі = ' + str(n))
    ss = 0
    for i in range(len(data)):
        data.iloc[ss,2] = data.iloc[i,0]
        if(data.iloc[i,1] == 0):
            if(Tn > 0):
                Tn -= 1
        else:
            Tp += 1
        data.iloc[ss,3] = (Tp*100)/p
        data.iloc[ss,4] = (Tn*100)/n
        if(i < len(data)-1 and data.iloc[i,0] == data.iloc[i+1,0]):
            continue
        ss+=1
    data['Se+Sp'] = [ 0 for i in range(len(data))]
    data['|Se-Sp|'] = [ 0 for i in range(len(data))]
    data['>80'] = [ 0 for i in range(len(data))]
    for i in range(ss):
        data.iloc[i,5] = data.iloc[i,3] + data.iloc[i,4]
        data.iloc[i,6] = abs(data.iloc[i,3] - data.iloc[i,4])
        data.iloc[i,7] = data.iloc[i,4] if(data.iloc[i,4] > 80) else 0
    kr1 = max(data['Se+Sp'][0:ss])
    kr2 = min(data['|Se-Sp|'][0:ss])
    kr3 = [data.iloc[i,3] if(data.iloc[i,7] != 0 ) else 0 for i in range(ss)]
    kr3 = data.iloc[kr3.index(max(kr3)),2]
    for i in range(ss):
        if(data.iloc[i,5] == kr1):
            kr1 = data.iloc[i,2]
        if(data.iloc[i,6] == kr2):
            kr2 = data.iloc[i,2]
        
    print('Критерій 1 = ' + str(kr1))
    print('критерій 2 = ' + str(kr2))
    print('критерій 3 = ' + str(kr3))
    fig = plt.figure(figsize=(8,8), dpi=120)
    ax = fig.add_axes([0,0,1,1])
    ax.set_xlabel('100 - sp', fontsize=25)
    ax.set_ylabel('se', fontsize=25)
    ax.plot([i for i in range(100)],[i for i in range(100)])
    ax.plot([0 for i in range(100)],[i for i in range(100)])
    ax.plot([i for i in range(100)],[100 for i in range(100)])
    
    x = [100 - data['sp'].iloc[i] for i in range(ss)]
    y = [data['se'].iloc[i] for i in range(ss)]
    ax.plot(x,y)

    AUC = 0
    for i in range(ss-1):
        AUC += (y[i+1] + y[i])*((x[i+1] - x[i])/2)
    print('AUC = ' + str (AUC/10000))
    
    print(data)
    return [kr1,kr2,kr3]
```
