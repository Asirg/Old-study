```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style("whitegrid")
```

<p style="color:#4ca8ed">
    <span style="margin-left:50px;font-size:150%">
        Завантаження вбудованих наборів даних:
    </span>
</p>

```python
tips = sns.load_dataset('tips')
tips.head(10)
```

<p style="color:#4ca8ed">
    <span style="font-size:200%">
        Візуалізація даних.
    </span>
    <br>
    <span style="margin-left:50px;font-size:150%">
        Графіки для аналізу кількісних даних:
    </span>
</p>

<p style="color:#4ca8ed">
    <span style="font-size:130%">
        1.displot. Гістограма та графік щільності росподілу
    </span>
</p>

```python
tips1 = tips.head(20)
sns.distplot(tips1['total_bill'],  rug=True, kde_kws={"shade": True}, bins= 6)
```

<p style="color:#4ca8ed">
    <span style="font-size:130%">
        2.jointplot. Створює діаграму розсіювання, за допомогою якої можна співставити розподіл двох змінних та їх
взаємозв’язок.
    </span>
</p>

```python
sns.jointplot(x = "total_bill", y = "tip", data=tips, kind="kde", color="#4ca8ed", height = 7, xlim = (0,50), ylim = (0,7))
#kind - задає тип. Можливі варіанти: scatter, reg, resid, kde, hex.
```

<p style="color:#4ca8ed">
    <span style="font-size:130%">
        3.pairplot – будує попарні взаємозв’язки між змінними в повному наборі даних.
    </span>
</p>

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue ="sex", palette = "Set1")
```

<p style="color:#4ca8ed">
    <span style="margin-left:50px;font-size:150%">
        Графіки для аналізу категоріальних даних:
    </span>
</p>

<p style="color:#4ca8ed">
    <span style="font-size:130%">
        1.barplot – дозволяють отримувати сукупні дані по кількісним даним в залежності від категоріальної змінної.
    </span>
</p>

```python
sns.barplot(x = "sex", y = "tip", data = tips1)
```

```python
tips2 = tips.head(10)
sns.barplot(x = "sex", y = "tip", data = tips2, estimator=len)
```

```python
sns.barplot(x = "day", y = "tip", data = tips,palette="rocket", estimator=np.median)
```

<p style="color:#4ca8ed">
    <span style="font-size:130%">
        2.boxplots - використовується для візуалізації розподілу категоріальних даних за певною кількісною змінною.
    </span>
</p>

```python
sns.boxplot(x = "day", y = "total_bill", data = tips, palette="pastel")
```

```python
sns.boxplot(x = "day", y = "total_bill",hue="smoker", data = tips, palette="pastel")
```

```python
sns.boxplot(data = tips, palette="pastel", orient='h')
```

<p style="color:#4ca8ed">
    <span style="font-size:130%">
        3.stripplot – представляє собою діаграму розсіювання, де одна за змінних категоріальна.
    </span>
</p>

```python
sns.stripplot(x = "day", y = "total_bill",hue="smoker", data = tips, palette="pastel", size = 6, linewidth = 1)
```

```python
sns.stripplot(x ="day", y = "total_bill", hue ="smoker",dodge=True, data = tips, palette="pastel", alpha=.35, size = 30)
```

<p style="color:#4ca8ed">
    <span style="margin-left:50px;font-size:150%">
        Графіки для аналізу взаємозв’язку даних:
    </span>
</p>

<p style="color:#4ca8ed">
    <span style="font-size:130%">
        Heatmap – графік взаємозв’язку (кореляції) між даними у вигляді cматриці.
    </span>
</p>

```python
f, ax = plt.subplots(figsize=(15, 8))
ax = sns.heatmap(tips.corr(), annot = True, linewidths=.5, cmap="YlGnBu")
```

<p style="color:#4ca8ed">
    <span style="margin-left:50px;font-size:150%">
        Графіки для відображення даних у вигляді сітки:
    </span>
</p>

```python
g = sns.FacetGrid(tips, col = "time", row = "smoker", hue = "day")
g = g.map(plt.hist, "total_bill")
g.add_legend(title="legend")
```
