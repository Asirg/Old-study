```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

<p style="color:#4ca8ed"><span style="margin-left:50px;font-size:150%">1.Створення Векторів.</span>
    <br>Вектор <b>x</b> - розміром 11, і зі значеннями 0...5.
    <br>Вектор <b>y</b>, де <b>y</b> функія: <b>y</b> = x^2
</p>

```python
x = np.linspace(0,5,11)
y = x**2
x
```

```python
y
```

<p style="color:#4ca8ed"><span style="margin-left:50px;font-size:150%">2.Побудова графіків.</span>
    <br><i>2.1. Функціональний метод.</i>
</p>

```python
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title")
```

<p style="color:#4ca8ed">Створення декілька графіків, <i>використовуючи subplot</i></p>

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Graph 1")

plt.subplot(1,2,2)
plt.plot(y,x,'b')
plt.xlabel("Y label")
plt.ylabel("X label")
plt.title("Graph 2")
```

<p style="color:#4ca8ed"><i>2.2.Об’єктно-орієнтований метод.</i></p>

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.set_xlabel("X label")
axes.set_ylabel("Y label")
axes.set_title("Graph")
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.1,0.5,0.4,0.3])

axes1.plot(x,y,'r')
axes1.set_xlabel("X label")
axes1.set_ylabel("Y label")
axes1.set_title("Graph 1")

axes2.plot(y,x,'b')
axes2.set_xlabel("Y label")
axes2.set_ylabel("X label")
axes2.set_title("Graph 2")
```

<p style="color:#4ca8ed">Створення декількох графіків, <i>використовуючи subplots</i></p>

```python
fig, axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y,'r')
axes[0].set_xlabel("X label")
axes[0].set_ylabel("Y label")
axes[0].set_title("Graph 1")

axes[1].plot(y,x,'b')
axes[1].set_xlabel("Y label")
axes[1].set_ylabel("X label")
axes[1].set_title("Graph 2")
```

<p style="color:#4ca8ed"><span style="margin-left:50px;font-size:150%">3.Редагування графіків.</span>
    <br>Додавання легенди на графік.
</p>

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label="x^2")
ax.plot(x,x**3,label="x^3")

ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_title("Graph")
ax.legend()
```

<p style="color:#4ca8ed">Змінення різних характеристик графіку.</p>

```python
fig = plt.figure(figsize=(15,5), dpi=120)
ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**3,label="x^3", 
        color="g",                                           #Задання кольору графіка
        lw = 2, ls ='--',                                    #Задання ширину лінії і типу
        marker='*', markersize=20, markerfacecolor = 'pink') #Задання типу маркера, його розміру та кольору
ax.plot(x,x**4,label="x^4", color="black")
ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_title("Graph")
ax.legend()
```
